```python
import pyspark
import pandas as pd
from pyspark.sql import SQLContext, SparkSession, functions as F
from pyspark.sql.functions import hour, when, col, date_format, to_timestamp
from pyspark.sql.functions import *
from pyspark.sql.types import *



sc = pyspark.SparkContext(appName="CSE_6242_Project")
sqlContext = SQLContext(sc)
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("Spark") \
    .getOrCreate()
```

```python
def load_data(path):
    df = sqlContext.read.option("header",True).csv(path)
#     df.printSchema()
    return df


# Load Files
GDP = load_data("US_GDP_From_1947.csv")
globalCCI = load_data("Global_CCI_From_1960.csv")
monthlyCPI = load_data("Monthly_CPI_from_1913.csv")
retailSales = load_data("US_Advance_Retail_Sales_From_1992.csv")
treasuryYield = load_data("US_Daily_10_Year_Treasury_Yield.csv")
GNP = load_data("US_GNP_From_1947.csv")
inflationRate = load_data("US_Inflation_Rates_From_1914.csv")
PCE = load_data("US_Monthly_Personal_Consumption_Expenditures_From_1959.csv")
nonFarmEmployees = load_data("US_Non_farm_Employees_From_1939.csv")
rentIndex = load_data("US_Rent_of_Primary_Residence_From_1914.csv")
balticIndex = load_data("baltic_Dry_Index_From_1985.csv")
gasPrice = load_data("US_Weekly_Gas_Prices_From_1993.csv")


df = load_data("stock_data.csv")

# Cast columns to correct types
# df = df.drop("_c0")
df = df.withColumn("Date",df["Date"].astype('Timestamp'))
df = df.withColumn("Open",df["Open"].astype('float'))
df = df.withColumn("High",df["High"].astype('float'))
df = df.withColumn("Low",df["Low"].astype('float'))
df = df.withColumn("Close",df["Close"].astype('float'))
df = df.withColumn("Volume",df["Volume"].astype('float'))
df = df.withColumn("Dividends",df["Dividends"].astype('float'))
df = df.withColumn("Stock Splits",df["Stock Splits"].astype('float'))
df = df.withColumn("Adj Close",df["Adj Close"].astype('float'))


# Combine quarter GDP
df = df.withColumn("year", year("date"))\
        .withColumn("quarter", quarter("date"))\
        .withColumn("month", month("date"))\
        .withColumn("week", weekofyear("date"))
GDP = GDP.withColumn("year", year("date"))\
        .withColumn("quarter", quarter("date"))\
        .withColumn("GDP", col("GDP").astype("float"))\
        .drop("Date")
df = df.join(GDP, ["year", "quarter"], how="left")


# Combine monthly CCI
CCI = globalCCI.select("Time","Value")\
        .where(col("location") == "USA")\
        .withColumnRenamed("Value","CCI")\
        .withColumn("year", year("time"))\
        .withColumn("month", month("time"))\
        .withColumn("CCI", col("CCI").astype("float"))\
        .drop("Time")
df = df.join(CCI, ["year", "month"], how="left")


# Combine Retail Index
retailSales = retailSales.withColumn("Retail", retailSales.RSXFS.astype("float"))\
        .withColumn("Date", retailSales["Date"].astype("Timestamp"))\
        .withColumn("year", year("date"))\
        .withColumn("month", month("DATE"))\
        .drop("RSXFS","Date")
df = df.join(retailSales, ["year", "month"], how="left")


# Combine Treasury Yield
treasuryYield = treasuryYield.withColumn("date", to_timestamp("date", "M/d/yy"))\
        .withColumn("date", expr('''IF(date > "2022-11-02", add_months(date,-1200), date)'''))\
        .withColumn("Yield", treasuryYield[" value"].astype("float"))\
        .drop(" value")\
        .orderBy("Date")

df = df.join(treasuryYield, "Date", how="left")

# Combine quarter GNP
GNP = GNP.withColumn("year", year("date"))\
        .withColumn("quarter", quarter("date"))\
        .withColumn("GNP", col("GNP").astype("float"))\
        .drop("Date")
df = df.join(GNP, ["year", "quarter"], how="left")

# Combine Monthly PCE
PCE = PCE.withColumn("date", PCE["date"].astype("Timestamp"))\
        .withColumn("year", year("date"))\
        .withColumn("month", month("date"))\
        .withColumn("PCEPI", col("PCEPI").astype("float"))\
        .drop("date")
df = df.join(PCE, ["year", "month"], how="left")

# Combine Monthly Non-Farm Employees
nonFarmEmployees = nonFarmEmployees.withColumn("date", nonFarmEmployees["date"].astype("Timestamp"))\
        .withColumn("year", year("date"))\
        .withColumn("month", month("date"))\
        .withColumn("PAYEMS", col("PAYEMS").astype("int"))\
        .drop("date")
df = df.join(nonFarmEmployees, ["year", "month"], how="left")

# Combine Monthly Rent Index
rentIndex = rentIndex.withColumn("date", rentIndex["date"].astype("Timestamp"))\
        .withColumn("year", year("date"))\
        .withColumn("month", month("date"))\
        .withColumn("CUUR0000SEHA", col("CUUR0000SEHA").astype("float"))\
        .withColumnRenamed("CUUR0000SEHA","Rent_Index")\
        .drop("date")
df = df.join(rentIndex, ["year", "month"], how="left")

# Combine Weekly Baltic Index
balticIndex = balticIndex.withColumn("Price",F.regexp_replace('Price',',','').astype('float'))\
        .select("Date", "Price")\
        .withColumn("date", to_timestamp("date", "MM/dd/yyyy"))\
        .withColumn("year", year("date"))\
        .withColumn("week", weekofyear("date"))\
        .withColumnRenamed("Price","Baltic_Index")\
        .drop("date")
df = df.join(balticIndex, ["year", "week"], how="left")

# Melt Monthly CPI with Pandas

tmp = monthlyCPI.select("*").toPandas().melt(id_vars=["Year"],value_vars=["Jan", "Feb", "Mar", "Apr", "May", "June",\
    "July", "Aug", "Sep", "Oct", "Nov", "Dec"],var_name="Month", value_name="CPI")
tmpSchema = StructType([StructField("Year", StringType(), False),\
                       StructField("Month", StringType(), False),\
                       StructField("CPI", StringType(), True)])

# Combine Monthly CPI
monthlyCPI = spark.createDataFrame(tmp, schema=tmpSchema) 
monthlyCPI = monthlyCPI.withColumn("Year", monthlyCPI["Year"].astype("Integer"))\
        .withColumn("Month", col("Month").substr(0,3))\
        .withColumn("Month", from_unixtime(unix_timestamp(col("Month"),'MMM'),'MM').astype("Integer"))\
        .withColumn("CPI", monthlyCPI["CPI"].astype("Float"))\
        .drop("date")
df = df.join(monthlyCPI, ["year", "month"], how="left")

#Combine Weekly Gas Price
gasPrice = gasPrice.withColumn("Date", from_unixtime(unix_timestamp(col("Date"),'MMM dd, yyyy'),'yyyy-MM-dd').astype("Timestamp"))\
    .withColumn("gas price", col("gas price").astype("float"))\
    .withColumn("year", year("date"))\
    .withColumn("week", weekofyear("date"))\
    .withColumnRenamed("gas price","Gas_Price")\
    .drop("Date")
df = df.join(gasPrice, ["year", "week"], how="left")


# Melt Inflation Rate with Pandas
tmp = inflationRate.select("*").toPandas().melt(id_vars=["Year"],value_vars=["Jan", "Feb", "Mar", "Apr", "May", "Jun",\
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],var_name="Month", value_name="CPI")
tmpSchema = StructType([StructField("Year", StringType(), False),\
                       StructField("Month", StringType(), False),\
                       StructField("Inflation_Rate", StringType(), True)])

inflationRate = spark.createDataFrame(tmp, schema=tmpSchema) 

# Combine Inflation Rate
inflationRate = inflationRate.withColumn("Year", inflationRate["Year"].astype("Integer"))\
        .withColumn("Month", col("Month").substr(0,3))\
        .withColumn("Month", from_unixtime(unix_timestamp(col("Month"),'MMM'),'MM').astype("Integer"))\
        .withColumn("Inflation_Rate", inflationRate["Inflation_Rate"].astype("Float"))\
        .drop("date")
df = df.join(inflationRate, ["year", "month"], how="left")


df.show(10)
df.printSchema()
# df.write.csv("stock_date_combined_with_index")
```
