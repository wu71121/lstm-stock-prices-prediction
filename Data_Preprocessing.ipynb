```python
import csv
import glob
import pandas as pd
import numpy as np
```

```python
data_frames = []
for file in glob.glob('Data/stock_date_combined_with_index/*.csv'):
    print(file)
    csvFile = pd.read_csv(file)
    data_frames.append(csvFile)
data = pd.concat(data_frames)


to_drop = ['Open','High','Low','_c0','Adj Close','Dividends','Volume','Stock Splits','month','quarter']
data = data.drop(columns=to_drop)

data = data.dropna(axis = 0)
print(data.info())
```

```python
def assign_unique_id_to_object_column(df,col):
    ref = df[[col,'CPI']].groupby(by=[col]).sum().reset_index()
    ref[col+'_id'] = ref.index
    ref=ref.drop(['CPI'],axis=1)
    ref.to_csv("./col_%s_to_%s_id.csv"%(col,col))
    df = df.join(ref.set_index(col), on=col)
    df = df.drop(columns=[col])

    return df

to_assign_id = ['company','sector','country']
for col in to_assign_id:
    print(col)
    data = assign_unique_id_to_object_column(data,col)


print(data.info())
data.head(10)
```
